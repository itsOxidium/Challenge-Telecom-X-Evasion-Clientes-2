##Encoding

###Utilizando Dummies

```python
#Usamos get_dummies para codificar
import pandas as pd

# Codificar variables categ√≥ricas
#Esto eliminar√° la primera categor√≠a de cada variable para evitar multicolinealidad (por eso usamos drop_first=True).

df_encoded_Dummy = pd.get_dummies(df_final_cleaned, drop_first=True)

# Verificar resultado
print("‚úÖ Dataset codificado:")
print(df_encoded_Dummy.shape)
df_encoded_Dummy.head()


```

###Utilizando OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Definir columnas categ√≥ricas
categorical_cols = ['MultipleLines', 'InternetService', 'Contract', 'PaymentMethod']

# Crear el transformador
encoder = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(drop='first', sparse_output=False), categorical_cols)
    ],
    remainder='passthrough'  # Mantiene las dem√°s columnas sin transformar
)


# Aplicar transformaci√≥n
df_encoded_array = encoder.fit_transform(df_final_cleaned)

# Convertir a DataFrame
import numpy as np
encoded_feature_names = encoder.named_transformers_['onehot'].get_feature_names_out(categorical_cols)
all_feature_names = np.concatenate([encoded_feature_names, df_final_cleaned.drop(columns=categorical_cols).columns])

df_encoded_OHE = pd.DataFrame(df_encoded_array, columns=all_feature_names)

# Verificar resultado
df_encoded_OHE.head()
```

##Balanceo de Clases (opcional)

```python
# puede marcar la diferencia en la capacidad del modelo para detectar cancelaciones reales. Si ya verificaste que la clase Churn = 1 est√° subrepresentada, aplicar t√©cnicas de balanceo es una jugada estrat√©gica.
 #Opciones para balancear clases
#| T√©cnica       | Descripci√≥n breve |
#| Undersampling  | Reduce la clase mayoritaria para igualar la minoritaria |
#| Oversampling   | Duplica aleatoriamente ejemplos de la clase minoritaria |
#| SMOTE          | Genera ejemplos sint√©ticos de la clase minoritaria usando interpolaci√≥n |
#| Class Weights  | Ajusta el peso de cada clase en el modelo para penalizar errores en la minoritaria |

# Ver cu√°ntos valores nulos hay por columna
print(df_encoded_Dummy.isnull().sum())

# Eliminar filas con NaNs (opcional: puedes imputar si prefieres)
df_encoded_clean = df_encoded_Dummy.dropna()
#SMOTE---------------------------------------------------------------------------------
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# Separar variables
X = df_encoded_clean.drop('Churn', axis=1)
y = df_encoded_clean['Churn']

# Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42
)

# Aplicar SMOTE
smote = SMOTE(random_state=42)
X_train_bal, y_train_bal = smote.fit_resample(X_train, y_train)

# Confirmar balanceo

print("\n El Balanceo de Cargas \n")
print(y_train_bal.value_counts())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# üîπ Convertir a DataFrame escalado si no lo tienes a√∫n
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train_bal.columns)

# üîπ A√±adir la variable objetivo
X_train_scaled_df['Churn_Yes'] = y_train_bal.reset_index(drop=True)

# üîπ Seleccionar solo columnas num√©ricas continuas
# Excluye columnas binarias (0/1) y categ√≥ricas codificadas
numeric_cols = [col for col in X_train_scaled_df.columns if X_train_scaled_df[col].nunique() > 2]

# üîπ Filtrar el DataFrame
df_numeric = X_train_scaled_df[numeric_cols + ['Churn_Yes']]

# üîπ Calcular matriz de correlaci√≥n
corr = df_numeric.corr()

# üîπ Filtrar variables con correlaci√≥n significativa
target_var = 'Churn_Yes'
limiar = 0.2
variaveis_relevantes = corr.index[abs(corr[target_var]) >= limiar].tolist()

if target_var not in variaveis_relevantes:
    variaveis_relevantes.append(target_var)

# üîπ Crear matriz filtrada y m√°scara
corr_filtrada = corr.loc[variaveis_relevantes, variaveis_relevantes]
mascara = np.triu(np.ones_like(corr_filtrada, dtype=bool))

# üîπ Graficar heatmap optimizado
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr_filtrada,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    center=0,
    square=True,
    linewidths=0.5,
    cbar_kws={"shrink": 0.7},
    mask=mascara
)
plt.title(f'üîç Heatmap de correlaci√≥n (solo variables cuantitativas) ‚â• {limiar} con "{target_var}"')
plt.show()
```

###Calculo de VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

# üîπ Excluimos la variable objetivo
X_vif = df_model.drop(columns=['Churn_Yes'])

# üîπ Calculamos VIF para cada variable
vif_data = pd.DataFrame()
vif_data['Variable'] = X_vif.columns
vif_data['VIF'] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]

# üîç Mostramos resultados
print("üìê VIF por variable:")
print(vif_data.sort_values(by='VIF', ascending=False))
```

##Entrenamiento de modelo Base (Regresion Logistica)

###Modelo Completo

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler

# üéØ Separaci√≥n de datos
X = df_model.drop(columns=['Churn_Yes'])
y = df_model['Churn_Yes']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# üîÑ Escalado
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# üì¶ Modelo completo
model_full = LogisticRegression(max_iter=1000)
model_full.fit(X_train_scaled, y_train)

# üìä Evaluaci√≥n
y_pred_full = model_full.predict(X_test_scaled)
print("üîπ Modelo completo:")
print(confusion_matrix(y_test, y_pred_full))
print(classification_report(y_test, y_pred_full))
```

###Modelo Reducido (Sin varibles VIF Alto)

```python
# üßº Eliminamos variables con VIF > 10
high_vif_vars = vif_data[vif_data['VIF'] > 10]['Variable'].tolist()
X_reduced = X.drop(columns=high_vif_vars)

# üîÅ Divisi√≥n y escalado
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
    X_reduced, y, test_size=0.2, stratify=y, random_state=42
)
X_train_r_scaled = scaler.fit_transform(X_train_r)
X_test_r_scaled = scaler.transform(X_test_r)

# üì¶ Modelo reducido
model_reduced = LogisticRegression(max_iter=1000)
model_reduced.fit(X_train_r_scaled, y_train_r)

# üìä Evaluaci√≥n
y_pred_reduced = model_reduced.predict(X_test_r_scaled)
print("üîπ Modelo reducido (sin VIF alto):")
print(confusion_matrix(y_test_r, y_pred_reduced))
print(classification_report(y_test_r, y_pred_reduced))
```
